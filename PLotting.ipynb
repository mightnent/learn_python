```
import numpy as np
import matplotlib.pyplot as plt
#start stop step
x = np.arange(0,2.3,0.2)
y = np.arange(0,2.3,0.2)

X, Y = np.meshgrid(x, y)
u = -Y
v = X

fig, ax = plt.subplots(figsize=(7,7))
ax.quiver(X,Y,u,v)

# ax.xaxis.set_ticks([])
# ax.yaxis.set_ticks([])
ax.axis([-0.2, 2.3, -0.2, 2.3])
ax.set_aspect('equal')

plt.show()
```
